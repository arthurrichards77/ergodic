```python
import glob
```

```python
file_list = glob.glob('run_*.json')
```

```python
fn = file_list[0]
```

```python
import json
```

```python
with open(fn,'r') as f:
    data_dict = json.load(f)
```

```python
data_dict
```

```python
import pandas
```

```python
df = pandas.DataFrame(columns=[k for k in data_dict.keys()])
```

```python
df
```

```python
for fn in file_list:
    with open(fn,'r') as f:
        data_dict = json.load(f)
    df.loc[len(df)] = pandas.Series(data_dict)
```

```python
df['wave_count'] = [len(w) for w in df.wave_numbers]
```

```python
df.sort_values(['pdf','mesh','num_moves','method','wave_count'])
```

```python
wave_df = df[(df['method']=='revisit-fix0-fix1') & (df['mesh']=='square 8x8') & (df['num_moves']==16) & (df['pdf'].isin(['cosine-ridges-2','uniform']))]
wave_df.sort_values(['pdf','mesh','num_moves','method','wave_count'])
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
wave_df.sort_values(['wave_count']).groupby('pdf').plot(x='wave_count',y='solve_time',ax=ax)
```
